```python
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 
for ele in text:
    if ele in punc:
        text = text.replace(ele, "")
```

```python
text = text.lower()
```

# Finding N-grams

```python
token = nltk.word_tokenize(text)
unigrams = ngrams(token, 1)
bigrams = ngrams(token,2)
trigrams = ngrams(token,3)

list0 = Counter(unigrams)
list1 = Counter(bigrams)
list2 = Counter(trigrams)
```

```python
print(list0) #For a word
```

```python
print(list1) # For 2 word phrases
```

```python
print(list2) # For 3 word phrases
```

# Counting the total words

```python
def returnSum(myDict):
     
    list = []
    for i in myDict:
        list.append(myDict[i])
    final = sum(list)
     
    return final


total_count0 = returnSum(list0)
print(total_count0)

total_count1 = returnSum(list1)
print(total_count1)

total_count2 = returnSum(list2)
print(total_count2)
```

# Density for 1 word

```python
den0=[]

for i in list0:
    density = 1
    density = (list0[i]*(1/1192))*100
    den0.append(density)

print(den0)


# Density = (Nkr x (Nwp / Tkn)) x 100
# Where:
# Density = your keyword density
# Nkr = how many times you repeated a specific key-phrase
# Nwp = number of words in your key-phrase
# Tkn = total words in the analyzed text
```
